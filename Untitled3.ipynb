```python
# =====================================================================
# 📝 TASK 1: DATA CREATION & WRANGLING (COMPANY STANDARD)
# Domain: AI Tools Usage & Jobs Impact Analysis
# Tool Used: Python (Pandas, NumPy)
# =====================================================================

import pandas as pd
import numpy as np

# Setting seed for consistency (so data doesn't change on every run)
np.random.seed(42)

# simulating a corporate survey dataset of 50 working professionals
survey_data = {
    'User_ID': [f'EMP_AI_{i:03d}' for i in range(1, 51)],
    'Age_Group': np.random.choice(['Gen-Z (18-24)', 'Millennials (25-34)', 'Gen-X (35+)'], 50, p=[0.5, 0.35, 0.15]),
    'Primary_AI_Tool': np.random.choice(['ChatGPT', 'Gemini', 'Midjourney', 'GitHub Copilot', 'Claude'], 50, p=[0.4, 0.25, 0.15, 0.12, 0.08]),
    'Daily_Usage_Hours': np.random.choice([1, 2, 3, 4, 5], 50, p=[0.3, 0.4, 0.15, 0.1, 0.05]),
    'Job_Domain': np.random.choice(['Software Development', 'Content Writing', 'Graphic Design', 'Digital Marketing', 'Education'], 50),
    'Work_Efficiency_Improvement_%': np.random.choice([20, 30, 40, 50, 60, 70], 50, p=[0.1, 0.2, 0.3, 0.2, 0.1, 0.1]),
    'Satisfaction_Score': np.random.choice([1, 2, 3, 4, 5], 50, p=[0.05, 0.05, 0.2, 0.4, 0.3]),

    # Adding User Feedback/Reviews for Task 4 (Sentiment Analysis)
    'User_Feedback': np.random.choice([
        'AI saves a lot of my time, it is amazing!',
        'I love how ChatGPT helps me in writing code quickly.',
        'Midjourney generates beautiful designs, very helpful.',
        'It is okay, but sometimes gives wrong information.',
        'I am worried that AI might take over my job in future.',
        'The tool is too complex and difficult to understand.',
        'Excellent tool for brainstorming new ideas.'
    ], 50)
}

# Creating the corporate DataFrame
df = pd.DataFrame(survey_data)

# Data Wrapping/Formatting: Ensuring correct data types
df['Daily_Usage_Hours'] = df['Daily_Usage_Hours'].astype(int)
df['Work_Efficiency_Improvement_%'] = df['Work_Efficiency_Improvement_%'].astype(int)

# Saving the raw data into a CSV file (Just like downloading from a database)
df.to_csv('AI_Tools_Survey_Data.csv', index=False)

print("✅ TASK 1 COMPLETED: Professional AI Survey Dataset Created and Saved as 'AI_Tools_Survey_Data.csv'!")
print(f"Total Records Extracted: {df.shape[0]} rows and {df.shape[1]} columns.\n")

# Displaying the top 5 records to verify
df.head()
```

```python
# =====================================================================
# 🔍 TASK 2: EXPLORATORY DATA ANALYSIS (EDA) USING SQL
# Tool Used: Python + SQLite3 (SQL)
# =====================================================================

import sqlite3
import pandas as pd

# 1. Connecting to SQL Database
conn = sqlite3.connect('AI_Data_Analytics.db')
cursor = conn.cursor()

# 2. Converting our Task 1 DataFrame into an SQL Table
df.to_sql('ai_survey_table', conn, if_exists='replace', index=False)
print("✅ SQL Table 'ai_survey_table' is ready for Analysis!\n")
print("========================================================\n")


# --- [SQL QUERY 1]: Checking total number of rows ---
query_1 = "SELECT COUNT(*) FROM ai_survey_table;"
cursor.execute(query_1)
total_rows = cursor.fetchone()[0]
print(f"📊 [QUERY 1 OUTPUT]: Total Rows in Dataset = {total_rows}\n")


# --- [SQL QUERY 2]: Viewing top 5 records to understand structure ---
print("📋 [QUERY 2 OUTPUT]: First 5 Records from SQL Table:")
query_2 = "SELECT User_ID, Age_Group, Primary_AI_Tool, Job_Domain FROM ai_survey_table LIMIT 5;"
cursor.execute(query_2)
records = cursor.fetchall()
for row in records:
    print(row)
print("\n")


# --- [SQL QUERY 3]: Count of users for each AI Tool (Trending Tools) ---
print("🔥 [QUERY 3 OUTPUT]: Most Popular AI Tools Count:")
query_3 = """
SELECT Primary_AI_Tool, COUNT(*) as User_Count
FROM ai_survey_table
GROUP BY Primary_AI_Tool
ORDER BY User_Count DESC;
"""
cursor.execute(query_3)
tool_trends = cursor.fetchall()
for row in tool_trends:
    print(f"👉 Tool: {row[0]} | Used by: {row[1]} professionals")
print("\n")


# --- [SQL QUERY 4]: Average Daily Usage Hours by Job Domain ---
print("💻 [QUERY 4 OUTPUT]: Average Daily Usage Hours by Job Domain:")
query_4 = """
SELECT Job_Domain, AVG(Daily_Usage_Hours) as Avg_Hours
FROM ai_survey_table
GROUP BY Job_Domain
ORDER BY Avg_Hours DESC;
"""
cursor.execute(query_4)
domain_usage = cursor.fetchall()
for row in domain_usage:
    print(f"👉 Domain: {row[0]} | Average Hours: {row[1]:.2f} hours/day")
print("\n")


# Closing the database connection
conn.close()
print("========================================================")
print("✅ TASK 2 COMPLETED SUCCESSFULLY WITH SQL OUTPUTS!")
```
